```python
n = 10
sum = 0
i = 1

while i <= n:
    num = float(input(f"Enter element {i}: "))
    sum += num
    i += 1

avg = sum / n
print(f"Average of {n} numbers is: {avg:.2f}")
```

```python
import math
n = int(input("Enter the number of elements: "))
product = 1
for i in range(n):
    number = float(input("Enter number: "))
    product *= number
geometric_mean = math.pow(product, 1/n)
print(f"The geometric mean is: {geometric_mean}")
```

```python
import math

def calculate_area_heron(side1, side2, side3):
  
    s = (side1 + side2 + side3) / 2
    
    area = math.sqrt(s * (s - side1) * (s - side2) * (s - side3))
    return area


a = 5
b = 6
c = 7
triangle_area = calculate_area_heron(a, b, c)
print(f"The area of the triangle is {triangle_area:.2f}")
```

```python
#Create a data frame with details of 10 students and columns as Roll Number, Name,Gender, Marks1, Marks2, Marks3. 
```

```python
import pandas as pd
data = {
    'Roll Number': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ivy', 'Jack'],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'F', 'F', 'M'],
    'Marks1': [85, 78, 92, 67, 88, 75, 91, 80, 95, 70],
    'Marks2': [90, 82, 88, 72, 34, 79, 93, 84, 97, 68],
    'Marks3': [88, 76, 94, 70, 87, 73, 90, 82, 96, 71]
}

students_df = pd.DataFrame(data)


print(students_df)
```

```python
#Create a new column with total marks
```

```python
students_df['Total Marks'] = students_df['Marks1'] + students_df['Marks2'] + students_df['Marks3']

print(students_df)
```

```python
#Find the lowest marks in Marks1
```

```python
lowest_marks = students_df['Marks1'].min()
print(f"The lowest marks in Marks1: {lowest_marks}")
```

```python
#Find the Highest marks in Marks2 
```

```python
highest_marks = students_df['Marks2'].max()
print(f"The highest marks in Marks2: {highest_marks}")
```

```python
# Find the average marks in Marks3 
```

```python
average_marks = students_df['Marks3'].mean()
print(f"The average marks in Marks3: {average_marks}")
```

```python
#Find student name with highest average 
```

```python
students_df['Average Marks'] = students_df[['Marks1', 'Marks2', 'Marks3']].mean(axis=1)
top_student = students_df.loc[students_df['Average Marks'].idxmax(), 'Name']
print(f"Student with Highest Average Marks: {top_student}")
```

```python
#Find how many students failed in Marks2 (<40)
```

```python
failed_students = students_df[students_df['Marks2'] < 40]
num_failed = failed_students.shape[0]

print(f"Number of students who failed in Marks2: {num_failed}")
```
